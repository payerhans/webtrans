```python
from googletrans import Translator
```

### Syntax:
`translated = translator.translate('text zum übersetzeb', src='ko', dest='ja')`
- src – source language (default: auto)
- dest – destination language (default: en)
- origin – original text
- text – translated text
- pronunciation – pronunciation


```python
# Translator erzeugen:
translator = Translator()
```

```python
text = "    Send the message using HTML or plain text, or do not send any message through LiquidFiles."
```

```python
translated = translator.translate(text, dest='de')
```

```python
print(translated.text)
```

## Tokenisation of text into sentences
using  the ntlk toolkit

```python
#import nltk
text = ""
infile = "t"
with open(infile, 'r') as inf:
    for i in inf.readlines():
        text = text + i
```

```python
from nltk.tokenize import sent_tokenize
# make a list of sentences of text
sents = sent_tokenize(text)
```

```python
type(sents)
```

### Translate each element of list

```python
for e in sents:
    transl = translator.translate(e, dest='de')
    sleep(3)
    print(transl.text)
```
